```python
x = df.drop(columns="Price")
y = df["Price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.1,random_state=302)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
```

```python
ohe.fit(x[["name","company","fuel_type"]])
```

```python
column_train = make_column_transformer((OneHotEncoder(categories=ohe.categories_),["name","company","fuel_type"]),
                                       
                                       remainder="passthrough")
```

```python
linear = LinearRegression()
```

```python
pipe = make_pipeline(column_train,linear)
```

```python
pipe.fit(x_train,y_train)
```

```python
test_predict = pipe.predict(x_test)
```

```python
linearscore = r2_score(y_test,test_predict)
```

```python
linearscore
```

```python
scores = []
for i in range(1000):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,random_state=i)
    linear = LinearRegression()
    pipe = make_pipeline(column_train,linear)
    pipe.fit(x_train,y_train)
    linearpred = pipe.predict(x_test)
    scores.append(r2_score(y_test,linearpred))
```
